```python
# Q1

import math

def f(x):
    return math.sin(10*x) + math.cos(3*x)

start, end = 3, 6
num_points = 100000
step = (end - start) / num_points

x_vals = [start + i * step for i in range(num_points + 1)]
y_vals = [f(x) for x in x_vals]

roots = []
for i in range(len(y_vals) - 1):
    if y_vals[i] * y_vals[i+1] < 0:
        x_root = x_vals[i] - y_vals[i] * (x_vals[i+1] - x_vals[i]) / (y_vals[i+1] - y_vals[i])
        roots.append(round(x_root, 4))

roots = sorted(list(set(roots)))

print(f"Number of roots = {len(roots)}")
print(f"Roots: {', '.join(map(str, roots))}")
```

```python
# Q2

import math

def custom_sin(theta, terms=20):

    result = 0
    for n in range(terms):
        numerator = (-1) ** n * theta ** (2*n + 1)
        denominator = math.factorial(2*n + 1)
        result += numerator / denominator
    return result

x = math.pi

print("[Upper Limit N, Sum S_N]")
print("-------------------------")

final_value = None
for N in range(5, 101, 5):
    s = 0
    for i in range(N + 1):
        sqrt_term = math.sqrt((i * x) / 100)
        sin_arg = (i * x) / 100
        sin_term = custom_sin(sin_arg)
        s += sqrt_term * sin_term
    
    final_value = s
    print(f"N = {N:3d} : S = {s:.8f}")

print("-------------------------")
print(f"Final value of S (for N=100): {final_value:.8f}")
print()
```

```python
# Q3

def reconstruct_permutation(d: list[int]) -> list[int] | None:

    n = len(d) + 1
    if n < 1:
        return None

    relative_perm = [0] * n
    
    current_sum = 0
    for i in range(n - 1):
        current_sum += d[i]
        relative_perm[i+1] = current_sum
        
    min_P = min(relative_perm)
    shift_C = 1 - min_P
    
    perm = [p + shift_C for p in relative_perm]
    
    
    if len(perm) != n:
        return None 
        
    if max(perm) != n:
        return None

    if set(perm) != set(range(1, n + 1)):
        return None

    return perm

def generate_test_description(d, result):
    n = len(d) + 1
    
    if result is None:
        relative_perm = [0] * n
        current_sum = 0
        for i in range(n - 1):
            current_sum += d[i]
            relative_perm[i+1] = current_sum
        
        min_P = min(relative_perm)
        shift_C = 1 - min_P
        perm = [p + shift_C for p in relative_perm]
        
        if max(perm) != n:
            return f"Invalid: Range is too large {tuple(perm)} for n={n}"
        elif len(set(perm)) != len(perm):
            return f"Invalid: Duplicates {perm} for n={n}"
        else:
            return f"Invalid: Duplicates/Missing {perm} for n={n}"
    else:
        return f"Valid: perm={result} of 1..{n}"

if __name__ == "__main__":
    
    test_cases = [
        [1, 1, 1],
        [-1, -1, -1],
        [],
        [0, 1],
        [1, -1, 2],
    ]
    
    for d in test_cases:
        result = reconstruct_permutation(d)
        description = generate_test_description(d, result)
        
        d_str = str(d) if d else "[]"
        
        print(f"\nDifference Array d: {d_str} (n={len(d) + 1})")
        print(f"Description: {description}")
        
        if result is None:
            print("Result: None (Impossible)")
        else:
            print(f"Result: {result}")
```

```python
# Q4

def min_expression_value(expr):

    tokens = []
    for char in expr:
        if char.isdigit():
            tokens.append(int(char))
        elif char in ['+', '*']:
            tokens.append(char)
    
    def helper(tokens):
        if len(tokens) == 1:
            return tokens[0]
        
        min_val = float('inf')
        
        for i in range(1, len(tokens), 2):
            operator = tokens[i]
            left_tokens = tokens[:i]
            right_tokens = tokens[i+1:]
            
            left_min = helper(left_tokens)
            right_min = helper(right_tokens)
            
            if operator == '+':
                min_val = min(min_val, left_min + right_min)
            else:
                min_val = min(min_val, left_min * right_min)
        
        return min_val
    
    min_result = helper(tokens)
    return min_result

test_expressions = [
    "1+2*3",
    "2*3+4",
    "1+2+3",
    "2*3*4",
    "1+2*3*4",
    "1"
]

for expr in test_expressions:
    min_val = min_expression_value(expr)
    print(f"Expression: {expr}")
    print(f"Minimum value: {min_val}")
    print()
```

```python
# Q5

def longest_word_chain(words):
    words_set = set(words)
    memo = {}
    
    def dfs(word):
        if word in memo:
            return memo[word]
        
        max_chain = 1
        best_next = None
        
        for i in range(len(word)):
            prev_word = word[:i] + word[i+1:]
            
            if prev_word in words_set:
                chain_len, chain = dfs(prev_word)
                if chain_len + 1 > max_chain:
                    max_chain = chain_len + 1
                    best_next = prev_word
        
        if best_next:
            _, rest_chain = dfs(best_next)
            current_chain = rest_chain + [word]
        else:
            current_chain = [word]
        
        memo[word] = (max_chain, current_chain)
        return max_chain, current_chain
    
    max_length = 0
    best_chain = []
    
    for word in words:
        length, chain = dfs(word)
        if length > max_length:
            max_length = length
            best_chain = chain
    
    return max_length, best_chain

test_cases = [
    ["a", "ab", "abc", "abcd", "abcde"],
    ["abc", "abd", "abde", "ade", "ae", "1"],
    ["ba", "bca", "bda", "bdca"],
    ["ab", "ba", "bca", "bda", "bdca"],
]

for words in test_cases:
    length, chain = longest_word_chain(words)
    print(f"Words: {words}")
    print(f"Longest chain length: {length}")
    print(f"Chain: {' -> '.join(chain)}")
    print()
```

```python
# Q6

from collections import defaultdict

def group_equivalent_strings(words):

    def get_canonical(word):
        return ''.join(sorted(word))
    
    groups = defaultdict(list)
    
    for word in words:
        canonical = get_canonical(word)
        groups[canonical].append(word)
    
    return list(groups.values())

test_cases = [
    ["abc", "bca", "cab", "xyz", "zxy", "yxz"],
    ["ab", "ba"],
    ["aab", "aba", "baa"],
    ["abcd", "dcba", "cdab"],
]

for words in test_cases:
    groups = group_equivalent_strings(words)
    print(f"Words: {words}")
    print(f"Groups:")
    for group in groups:
        print(f"  {group}")
    print()
```

```python
# Q7

class NestedProduct:
    def __init__(self, spec):
        self.spec = spec
        self.factors = []
        
        for item in spec:
            if isinstance(item, list):
                self.factors.append(sorted(item))
            else:
                self.factors.append([item])
        
        self.factors = sorted(self.factors)
        self.current = [0] * len(self.factors)
        self.finished = len(self.factors) == 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.finished:
            raise StopIteration
        
        result = ''.join(str(self.factors[i][self.current[i]]) for i in range(len(self.factors)))
        
        i = len(self.current) - 1
        while i >= 0:
            self.current[i] += 1
            if self.current[i] < len(self.factors[i]):
                break
            self.current[i] = 0
            i -= 1
        
        if i < 0:
            self.finished = True
        
        return result

test_cases = [
    [['a', 'b'], ['1', '2', '3']],
    [['x', 'y'], ['p', 'q']],
    [['x', 'y'], ['x', 'y']],
    [['1', '2'], ['a', 'b', 'c'], ['x', 'y']],
    ['a', ['1', '2']],
]

for spec in test_cases:
    print(f"Specification: {spec}")
    products = list(NestedProduct(spec))
    print(f"Cartesian product: {products}")
    print()
```

```python
# Q8

def next_permutation(arr: list) -> bool:
    n = len(arr)
    if n <= 1:
        return False

    p = n - 2
    while p >= 0 and arr[p] >= arr[p + 1]:
        p -= 1

    if p < 0:
        arr.reverse()
        return False

    s = n - 1
    while arr[s] <= arr[p]:
        s -= 1
        
    arr[p], arr[s] = arr[s], arr[p]

    left, right = p + 1, n - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
        
    return True

print("Test Case 1: [1, 2, 3]")
arr = [1, 2, 3]
print(f"Start: {arr}")

result = next_permutation(arr)
print(f"Next: {result}, Array: {arr}")

while result:
    result = next_permutation(arr)
    print(f"Next: {result}, Array: {arr}")

print("\nTest Case 2: [1, 1, 2]")
arr = [1, 1, 2]
print(f"Start: {arr}")

result = next_permutation(arr)
print(f"Next: {result}, Array: {arr}")

while result:
    result = next_permutation(arr)
    print(f"Next: {result}, Array: {arr}")

print("\nTest Case 3: [2, 3, 1]")
arr = [2, 3, 1]
print(f"Start: {arr}")

result = next_permutation(arr)
print(f"Next: {result}, Array: {arr}")

while result:
    result = next_permutation(arr)
    print(f"Next: {result}, Array: {arr}")

print("\nTest Case 4: [1, 2, 2, 3]")
arr = [1, 2, 2, 3]
print(f"Start: {arr}")

result = next_permutation(arr)
print(f"Next: {result}, Array: {arr}")

while result:
    result = next_permutation(arr)
    print(f"Next: {result}, Array: {arr}")
```
